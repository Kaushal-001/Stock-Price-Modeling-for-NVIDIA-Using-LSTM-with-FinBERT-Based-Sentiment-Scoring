```python
from newsapi import NewsApiClient
import datetime as dt
import pandas as pd

news_api = NewsApiClient(api_key = "697564cc5c1d4480bf771c37bfcf8996")

#Define keywords
keywords = [
    "NVIDIA stock", "NVIDIA shares", "NVIDIA earnings", "NVIDIA investment",
    "NVIDIA AI chip", "NVIDIA GPU", "NVIDIA data center", "NVIDIA revenue",
    "NVIDIA financial results", "NVIDIA forecast"
]

today = dt.date.today()
from_date = today - dt.timedelta(days=30)

#containers for all article
all_articles = []

#Fetch article for each keywords
for kw in keywords:
  data = news_api.get_everything(
      q = kw,
      language='en',
      from_param=from_date.isoformat(),
      to=today.isoformat(),
      sort_by='relevancy',
      page_size=10
  )

  articles = data.get('articles',[])
  all_articles.extend(articles)


df = pd.DataFrame(all_articles)

```

```python
df = df.drop_duplicates(subset = 'url')
```

```python
df = df.sort_values(by='publishedAt', ascending=False).reset_index(drop=True)
```
